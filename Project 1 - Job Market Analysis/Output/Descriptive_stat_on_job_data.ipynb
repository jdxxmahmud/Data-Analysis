### Types of variable analysis
1. Univariate (1 feature)
2. Bivariate (2 features)
3. Multivariate (>2 features)


##### Univariate
1. Type of available data
2. Where the distribution of data looks correct or not.
3. Anomaly exists or not.

Univariate can be done both on **Continuous data** and **Categorical Data**

> For **Univariate Continuous** data, we use: <br>
> Boxplot, Histrogram, Distribution Plot (Graphic Analysis) <br>
> Describe, (Tabular Analysis)


> For **Univariate Categorical** data, we use: <br>
> Frequency Distribution(value_counts function), Bar Chart, Pie Chart, Countplot(seaborn module)

#### Bivariate

> Two features, x and y

##### There can be 3 scenarios
1. X => Categorical, Y => Numerical  ===> Pie Chart, Donut Chart, Bubble, Treemap.
2. X => Num, Y=> Num ===> Correlation, Scatterplot, Regression (Provided x and y are correlated.)
3. X=> Cat, Y=> Num ===> Chi-squared test
