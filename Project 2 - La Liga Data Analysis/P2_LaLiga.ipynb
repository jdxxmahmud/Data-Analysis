```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
temp_list = df['Season'][0].split('-')
temp_list[1] = ('').join(['19', temp_list[1]])
# temp_list[1] = f"19{temp_list[1]}"
temp_list
```

```python
season_timeline = [[], []]
for i in df['Season']:
    temp_list = i.split('-')
    season_timeline[0].append(temp_list[0])
    season_timeline[1].append(('').join(['19', temp_list[1]]))

# season_timeline[1]
```

```python
df.drop(['Start Year', 'End Year'], axis = 1, inplace = True)
df.insert(1, column='Start Year', value=season_timeline[0])
df.insert(2, column='End Year', value=season_timeline[1])

df[['Start Year', 'End Year']] = df[['Start Year', 'End Year']].apply(pd.to_numeric)

df.head()
```
